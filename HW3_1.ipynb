Recall that,

Written K-strike straddle is a portfolio of

- selling K-strike call of one unit
- selling K-strike put of one unit

Butterfly with three stikes $K_1 < K_2 < K_3$ is the portfolio of

- 1 unit of written K-strike straddle
- 1 unit of purchased K-strike call
- 1 unit of purchased K-strike put

Plot a diagram of exercise price versus payoff for the following portfolios:

- written 40-strike straddle
- a butterfly consists of
  - written 40-strike straddle
  - purchased 45-strike call
  - purchased 35-strike put

```
class VanillaOption:
    def __init__(
        self,
        otype = 1, # 1: 'call'
                  # -1: 'put'
        strike = 40.,
        maturity = 1.,
        market_price = 10.):
      self.otype = otype
      self.strike = strike
      self.maturity = maturity
      self.market_price = market_price #this will be used for calibration

    def buy_payoff(self, s):
      otype = self.otype
      k = self.strike
      maturity = self.maturity
      return max([0, (s - k)*otype])

        
    def sell_payoff(self, s): #s: excercise price
      otype = self.otype
      k = self.strike
      maturity = self.maturity
      return -max([0, (s - k)*otype])

import numpy as np
import matplotlib.pyplot as plt
o1= VanillaOption(otype = 1) 
o2= VanillaOption(otype = -1)
ex_pri = range(20,61)
payoff = [o1.sell_payoff(i)+o2.sell_payoff(i) for i in ex_pri]
#draw the graph#
plt.plot(ex_pri, payoff)
plt.xlabel('exercise price')
plt.ylabel('straddle_payoff')
plt.show()
o3 = VanillaOption(otype = 1, strike = 45) 
o4= VanillaOption(otype = -1, strike = 35)
butter_payoff = [o1.sell_payoff(i)+o2.sell_payoff(i)+o3.buy_payoff(i)+o4.buy_payoff(i) for i in ex_pri]
#draw the graph#
plt.plot(ex_pri, butter_payoff)
plt.xlabel('exercise price')
plt.ylabel('butterfly_payoff')
plt.show()


  
```
